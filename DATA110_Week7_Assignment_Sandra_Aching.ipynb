# Focused Iris Data Analysis Assignment

## Objective
In this assignment, you will perform specific data visualization and analysis tasks on the Iris dataset similar to what we did in week 7 related to Penguins Dataset
## Dataset
You will be working with the 'Iris' dataset, which is available in the seaborn library. This dataset contains measurements of iris flowers, including sepal length, sepal width, petal length, petal width, and species.






```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load the iris dataset from seaborn
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris.head())
```

## Tasks

### 1. Data Visualization

#### 1.1 Pair Plot
Create a pair plot of all features in the Iris dataset, color-coding by species. Include an overall title for the plot.


```python
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.title("Pairplot for all features by Species")

plt.show()
```

#### 1.2 Bar Plot
Create a bar plot showing the average sepal length for each iris species.

```python
sns.barplot(x="species", y="sepal_length", data=iris)
plt.title("Average Sepal Length by Species")
plt.show()
```

#### 1.3 Box Plot
Create a box plot showing the distribution of petal widths for each species.

```python
sns.boxplot(x="species", y="petal_width", data=iris)
plt.title('Petal Width Distribution by Species')
plt.show()
```

#### 1.4 Grouped Bar Plot
Create a grouped bar plot showing the average sepal and petal lengths by species.

```python
# Grouped bar plot: Sepal Length vs Petal Length, colored by species
plt.figure(figsize=(30, 20))
sns.barplot(x='sepal_length', y='petal_length', hue='species', data=iris)

# Add title and labels
plt.title('Average Sepal Length and Petal Lengths by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

# Show the plot
plt.show()
```

#### 1.5 Violin Plot
Create a violin plot showing the distribution of sepal length for each species.


```python
# Create the violin plot
sns.violinplot(x='species', y='sepal_length', data=iris, inner=None, color="skyblue")

# Customizing the plot
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')

plt.show()
```

### 2. Data Analysis

#### 2.1 Aggregation
a) Calculate and display the mean, minimum, and maximum values of all numerical features (sepal length, sepal width, petal length, petal width) for each species.

b) Count and display the number of samples in each species.


```python
#Mean values of all numerical features
species_sepal_length_means = iris.groupby('species')['sepal_length'].mean()
print(species_sepal_length_means)
species_sepal_width_means = iris.groupby('species')['sepal_width'].mean()
print(species_sepal_width_means)
species_petal_length_means = iris.groupby('species')['petal_length'].mean()
print(species_petal_length_means)
species_petal_width_means = iris.groupby('species')['petal_width'].mean()
print(species_petal_width_means)

#Minimum values of all numerical features
species_sepal_length_min = iris.groupby('species')['sepal_length'].min()
print(species_sepal_length_min)
species_sepal_width_min = iris.groupby('species')['sepal_width'].min()
print(species_sepal_width_min)
species_petal_length_min = iris.groupby('species')['petal_length'].min()
print(species_petal_length_min)
species_petal_width_min = iris.groupby('species')['petal_width'].min()
print(species_petal_width_min)

#Maximum values of all numerical features
species_sepal_length_max = iris.groupby('species')['sepal_length'].max()
print(species_sepal_length_max)
species_sepal_width_max = iris.groupby('species')['sepal_width'].max()
print(species_sepal_width_max)
species_petal_length_max = iris.groupby('species')['petal_length'].max()
print(species_petal_length_max)
species_petal_width_max = iris.groupby('species')['petal_width'].max()
print(species_petal_width_max)



```
