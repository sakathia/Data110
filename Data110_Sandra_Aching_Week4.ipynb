## Task 1: Basic Bar Chart for Crime Rates in 2023

In this task, you will create a **bar chart** showing the offense rates per 100,000 people for each type of crime in 2023.

### Instructions:
- Create a bar chart for crime rates in 2023.
- Customize the bar colors and add gridlines.
- Label the axes and add a title.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Creating the dataframe for crime rates in 2023 from the table provided
df_crime_data = pd.read_csv('https://raw.githubusercontent.com/sakathia/Data110/refs/heads/main/dataset/Crimes_Data.csv')
df_crime_data
```

```python
df_crime_data.info()  # Provides information about the dataframe
```

```python
plt.figure(figsize=(8,6))
plt.bar(df_crime_data['Offense'], df_crime_data['Offenses per 100,000 People (2023)'], color='blue')
plt.xticks(rotation=45,fontsize=8)
plt.title('National Crime Trends for 2023', fontsize=16)
plt.xlabel('Offense')
plt.ylabel('Offenses per 100,000 People (2023)')
plt.grid(True)
plt.show()
```

## Task 2: Change in Crime Rates (2022 vs. 2023) with Unique Colors

In this task, you will create a **horizontal bar chart** that visualizes the **percentage change** in offense rates from 2022 to 2023 for each crime type. Each bar should have a unique color for better distinction.

### Instructions:
- Create a horizontal bar chart showing the **change in offense rates**.
- Assign a unique color to each crime type.
- Label the axes and add a title.

```python
plt.figure(figsize=(8, 6))
colors=['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'gray']
plt.barh(df_crime_data['Offense'], df_crime_data['Change in Offenses per 100,000 People (2022-2023)'], color=colors)
plt.title('Percentage Change in Offense Rates from 2022 to 2023 for each Crime Type', fontsize=14)
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)', fontsize=12)
plt.ylabel('Offense', fontsize=12)
plt.grid(True)
plt.show()
```

### Additional Instructions:
1. **Why does this bar graph make more sense than Task 1's bar graph?**
   - Think about how **absolute numbers** (like total crime rates) might be misleading compared to **percentage changes**.
   - Explain why seeing the **change** over time is more insightful than just seeing the total number of offenses for a single year.

Answer: While it is good to see the absolute numbers for the total crime rates, it can be difficult to understand the trends or patterns from year to year. With the total number for each crime type we are not able to tell if there is any improvement or if there was a change between any of the crime types. For example, in the first graph, we can see that Larceny was had the highest crime rate for 2023 but we do not know if this was the case over the years or if this is an anomaly. Therefore, it is more insightful to see the change over time to understand if there is any patterns or analomalies that we can see through different periods of time. In this case, it is more insightful to see the change over time to know if there was any improvement or decline in crime rates between years and by crime type.

2. **Improve the clarity of your bar graph**:
   - Create **multiple versions** of the bar chart for Task 2, each time making the chart more **clear and readable**.
   - Experiment with different color schemes, labels, and gridlines.
   - In each iteration, focus on improving the ease with which someone can quickly understand the trends in crime rates.

```python
plt.figure(figsize=(8, 6))
colors=['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'gray']
plt.barh(df_crime_data['Offense'], df_crime_data['Change in Offenses per 100,000 People (2022-2023)'], color=colors)
plt.title('Percentage Change in Offense Rates from 2022 to 2023 for each Crime Type', fontsize=14)
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)', fontsize=12)
plt.ylabel('Offense', fontsize=12)
plt.grid(axis='x', linestyle='-', color='gray')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
colors=['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'gray']
plt.barh(df_crime_data['Offense'], df_crime_data['Change in Offenses per 100,000 People (2022-2023)'], color=colors, zorder=3)
ax = plt.gca()
ax.spines['top'].set_visible(False) # Taking out the top border
ax.spines['right'].set_visible(False) #Taking out the right border
plt.title('Percentage Change in Offense Rates from 2022 to 2023 for each Crime Type', fontsize=14)
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)', fontsize=12)
plt.ylabel('Offense', fontsize=12)
plt.grid(axis='x', linestyle='-', color='#d3d3d3', zorder=0) # Putting horizontal gridline behind the bars
plt.show()
```

```python
plt.figure(figsize=(8, 6))
colors = ['darkred', 'darkorange', 'gold', 'darkgreen', 'darkblue', 'indigo', 'dimgray']   #different color scheme
plt.barh(df_crime_data['Offense'], df_crime_data['Change in Offenses per 100,000 People (2022-2023)'], color=colors, zorder=3)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.title('Percentage Change in Offense Rates from 2022 to 2023 for each Crime Type', fontsize=14)
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)', fontsize=12)
plt.ylabel('Offense', fontsize=12)
plt.grid(axis='x', linestyle='-', color='#d3d3d3', zorder=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
colors = ['darkred', 'darkorange', 'gold', 'darkgreen', 'darkblue', 'indigo', 'dimgray']
plt.barh(df_crime_data['Offense'], df_crime_data['Change in Offenses per 100,000 People (2022-2023)'], color=colors, zorder=3)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.title('Percentage Change in Offense Rates from 2022 to 2023 for each Crime Type', fontsize=14)
plt.xlabel('Change in Offenses per 100,000 People (2022-2023)', fontsize=12)
plt.ylabel('Offense', fontsize=12)
ax.tick_params(axis='x', labelcolor='#808080', labelsize= 9) # Changing the x label size and color
ax.tick_params(axis='y', labelcolor='#808080', labelsize= 9) # Changing the y label size and color
plt.grid(axis='x', linestyle='-', color='#d3d3d3', zorder=0)
plt.show()
```
