```python
#Installing the needed packages

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
moco_data=pd.read_csv('https://raw.githubusercontent.com/sakathia/Data110/refs/heads/main/dataset/Daily_Arrests_20241001%20(1).csv')
moco_data.head()
```

1. **Filter Theft-Related Offenses**
   - Filter the dataset to include only rows where the `OFFENSE` column contains the word "theft".
   - Create a new DataFrame called `theft_df` for this filtered data.


```python
theft_df=moco_data[moco_data['OFFENSE'].str.contains('theft',case=False,na=False)]
theft_df.head()
```

3. **Visualizing Theft Offenses by Day of the Week**
  
    
   - **Task**: Which day of the week has the highest number of theft arrests? What might be the reasons for this trend?


```python
theft_df['ARREST DATE']=pd.to_datetime(theft_df['ARREST DATE'])
theft_df['dayOfWeek']=theft_df['ARREST DATE'].dt.dayofweek
theft_df.info()
theft_df.describe()
```

```python
day_counts=theft_df['dayOfWeek'].value_counts()
# Map the day of the week index to its name
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.bar(day_counts.index,day_counts)
```

```python
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.barh(day_names ,day_counts)
plt.title('Theft Arrests by Day of the Week')
plt.show()
```

The day of the week with the most Theft arrests is on Monday. This actually I believe may be due to the offense occuring on the weekend and it offcially being reported and investigated on Monday leading to an arrest possibly on the same day.

4. **Age Distribution of Theft Offenders**
   

   - **Task**: Describe the age distribution. Which age group seems to be most involved in theft offenses?

```python
plt.hist(theft_df['AGE'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age for Theft Offenses')
plt.show()
```

```python
sns.kdeplot(theft_df['AGE'])
plt.title('Distribution of Age for Theft Offenses')
plt.show()
```

The Age distribution is right skewed. I can tell from the histogram that there are some variations in age groups for example for people in 30's it is much less theft offense arrests than people in the mid 20's age group. However, from the kde plot I can tell that despite the variations the density of the range for ages that have been arrested for theft offense falls between early 20's and 40's. The age group that seems to have the highest amount of theft arrests is people in the mid 20's age group.

5. **Offense Analysis**
   - Analyze the types of theft offenses.
   - **Task**: Use `value_counts()` on the `OFFENSE` column to find out what kinds of theft are most common. Create a bar chart to visualize this.


```python
theft_offense_counts=theft_df['OFFENSE'].value_counts()
plt.barh(theft_offense_counts.index,theft_offense_counts)
plt.title('Types of Theft Offenses')
plt.show()
```

The most common Theft Offense Arrest tends to be Theft of 100 dollars to under 1,500 dollars, Thefts less than 100 dollars and Thefts of 1,500 dollars to under 25,000 dollars.

6. Can you make some other visualizaion related to theft offenses?

```python
# I am visualizing the theft related offenses in a pie chart
plt.figure(figsize=(8, 8))
plt.pie(theft_offense_counts, labels=theft_offense_counts.index, autopct='%1.1f%%')
plt.title('Theft Related Offenses')
plt.show()
```
