### 2. Basic Line Plot
- Create a plot of the quadratic function $f(x) = x^2 - 2x + 1$ over the range $[-2, 4]$

### 3. Multiple Lines
- Create a plot with the following three functions on the same graph:
  - $f(x) = x^3$
  - $g(x) = e^x$
  - $h(x) = \log(x)$ (for $x > 0$)
- Use different colors for each line
- Add a legend to distinguish between the functions

```python
#Plot for Multiple Lines
x1 = np.linspace(-10, 10, 50)
x2 = np.linspace(0.1, 10, 50)  # For h(x) = log(x), needs x > 0

plt.plot(x1, x1**3, 'red', x1, np.exp(x1), 'blue', x2, np.log(x2), 'green')
plt.title('Plot of f(x) = x^3, g(x) = e^x, and h(x) = log(x)')

plt.legend(['x^3', 'e^x', 'log(x)'])
plt.show()
```

### 5. Axes Limits and Scaling
- Plot the function $f(x) = \frac{1}{x^2 + 1}$ over the range $[-10, 10]$
- Set appropriate custom x and y limits to showcase the function's behavior
- Create another plot of the same function with reversed x and y axes
- Demonstrate the use of `plt.axis('tight')` and `plt.axis('equal')`

```python
#plot using plt.axis('tight') and custom x and y limits
x = np.linspace(-10, 10, 100)

f = 1 / (x**2 + 1)

plt.plot(x, f, color='blue')
plt.title('Plot of f(x) = 1/(x^2 + 1)')
plt.xlim(-10, 10)  # Set x limits
plt.ylim(0, 1.1)   # Set y limits
plt.axis('tight')
```

```python
#plot using plt.axis('equal') and custom x and y axes limits reversed
x = np.linspace(-10, 10, 100)

f = 1 / (x**2 + 1)

plt.plot(x, f, color='blue')
plt.title('Plot of f(x) = 1/(x^2 + 1)')
plt.xlim(0, 1.1) # Set x limits
plt.ylim(-10, 10)  # Set y limits
plt.axis('equal')
plt.show()
```

### 6. Labels and Legends
- Create a plot of $f(x) = \sqrt{x}$ and $g(x) = x^{\frac{1}{3}}$ for $x \geq 0$

### 8. Customizing Scatter Plots
- Generate data for the function $f(x) = x \sin(x)$ over the range $[0, 4\pi]$
- Create a scatter plot of this data with a line connecting the points
- Customize the plot by changing the marker size, line width, marker face color, marker edge color, and marker edge width


```python
#Customizing Scatterplots
x = np.linspace(0, 4 * np.pi, 100)
y = x*np.sin(x)

plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='red',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.title('Plot of x * sin(x) over [0, 4pi]', fontsize=14)
plt.show()
```

### 9. Combined Plot
- Plot the function $f(x) = x \cos(x)$ over the range $[-2\pi, 2\pi]$
- Add scatter points for x values at every $\frac{\pi}{4}$ interval
- Use different colors and styles for the line and scatter components

```python
#Showing a combined plot
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = x * np.cos(x)
x_scatter = np.arange(-2 * np.pi, 2 * np.pi + np.pi / 4, np.pi / 4) #Allows to be able to plot points in increments of pi/4
y_scatter = x_scatter * np.cos(x_scatter)
plt.plot(x, y, color='blue', linestyle='--', linewidth=2)
plt.scatter(x_scatter, y_scatter, color='red', marker='o', s=70)
plt.title('Plot of f(x) = xcos(x) over [-2pi, 2pi]', fontsize=14)
plt.show()
```
